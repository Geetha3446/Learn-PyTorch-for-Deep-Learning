# Matrix multiplication

Let's create a tensor and perform element-wise multiplication and matrix multiplication on it.

```python
tensor = torch.tensor([1,2,3])
tensor.shape
```

```python
# Element-wise matrix multiplication
tensor * tensor
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
# Can also use the "@" symbol for matrix multiplication, though not recommended
tensor @ tensor
```

You can do matrix multiplication by hand but it's not recommended.

The in-built torch.matmul() method is faster.

```python
%%time

value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
value
```

#One of the most common errors in deep learning (shape errors)

```python
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]],dtype = torch.float32)
tensor_B = torch.tensor([[7,8],
                         [9,10],
                         [11,12]],dtype = torch.float32)
```

```python
print(tensor_A)
print(tensor_B)
```

```python
print(tensor_A)
print(tensor_B.T)
```

```python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

```python
torch.mm(tensor_A, tensor_B.T)
```

```python
# Since the linear layer starts with a random weights matrix, let's make it reproducible (more on this later)
torch.manual_seed(42)
# This uses matrix multiplication
linear = torch.nn.Linear(in_features=2, # in_features = matches inner dimension of input
                         out_features=6) # out_features = describes outer value
x = tensor_A
output = linear(x)
print(f"Input shape: {x.shape}\n")
print(f"Output:\n{output}\n\nOutput shape: {output.shape}")
```

#Reshaping, stacking, squeezing and unsqueezing

```python
y = torch.arange(1,7)
y, y.shape
```

Now let's add an extra dimension with torch.reshape().

```python
y_reshaped = y.reshape(1,6)
y, y.shape, y_reshaped, y_reshaped.shape
```

modify the view, original tensor also changes
why?
Because both point to same memory.

```python
z = y.view(2,3)
z
```

modify the view

```python
z[0][0] = 99
z
```

original tensor also changes
why?
Because both point to same memory.

```python
y
```

```python
# Stack tensors on top of each other
y_stacked = torch.stack([y, y, y, y], dim=0) # try changing dim to dim=1 and see what happens
y_stacked
```

```python
print(f"Previous tensor: {y_reshaped}")
print(f"Previous shape: {y_reshaped.shape}")

# Remove extra dimension from x_reshaped
y_squeezed = y_reshaped.squeeze()
print(f"\nNew tensor: {y_squeezed}")
print(f"New shape: {y_squeezed.shape}")
```

```python
print(f"Previous tensor: {y_squeezed}")
print(f"Previous shape: {y_squeezed.shape}")

## Add an extra dimension with unsqueeze
y_unsqueezed = y_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {y_unsqueezed}")
print(f"New shape: {y_unsqueezed.shape}")
```

```python
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

Because permuting returns a view (shares the same data as the original), the values in the permuted tensor will be the same as the original tensor and if you change the values in the view, it will change the values of the original.

# PyTorch tensors & NumPy

##NumPy Array → PyTorch Tensor

```python
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

torch.from_numpy() usually shares memory with NumPy array.

That means:

changing one can affect the other

BUT in your example:


```python
array = array + 1
```

does NOT modify original array in-place.

```python
array, tensor
```

##in-place modification

```python
array = np.array([1.,2.,3.])
tensor = torch.from_numpy(array)
array += 1
```

This WOULD affect tensor too.

Because memory is shared.

```python
array, tensor
```

##Tensor → NumPy Array

```python
tensor = torch.ones(7)

numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

Tensor and NumPy array usually share memory.

BUT your example:

```python
tensor = tensor + 1
```

creates a NEW tensor.

So NumPy array still points to old memory.

```python
tensor, numpy_tensor
```

inplace same as above

#Reproducibility (trying to take the random out of random)

```python
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

random_tensor_A, random_tensor_B
```

```python
print(f"Does Tensor A equal Tensor B? (anywhere)")
random_tensor_A == random_tensor_B
```

That's where torch.manual_seed(seed) comes in, where seed is an integer (like 42 but it could be anything) that flavours the randomness.

```python
import random

RANDOM_SEED=42
torch.manual_seed(seed=RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.random.manual_seed(seed=RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(f"Tensor C:\n{random_tensor_C}\n")
print(f"Tensor D:\n{random_tensor_D}\n")
print(f"Does Tensor C equal Tensor D? (anywhere)")
random_tensor_C == random_tensor_D
```
